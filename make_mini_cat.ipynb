```python
#i cant run bagpipes and get a whole new thing of galaxies because that would take hours and too much morgan
#so i am going to make a mini catalogue of 5 galaxies and use this to plot....

from astropy.io import fits
from astropy.table import Table
from astropy.table import join
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

hdulist = fits.open("/raid/scratch/work/austind/GALFIND_WORK/Catalogues/v13/ACS_WFC+NIRCam/JADES-DR3-GN/(0.32)as/JADES-DR3-GN_MASTER_Sel-F277W+F356W+F444W_v13.fits")
table_objects = Table(hdulist[4].data)
table_selection = Table(hdulist[6].data)
hdulist.close()
print(table_objects.colnames)
print(table_selection.colnames)

mask = table_selection['Austin+25_EAZY_fsps_larson_zfree_0.32as'] == True
table_galaxies = table_objects[mask]


# Randomly select N rows from the table
np.random.seed(42)  # optional: makes the selection reproducible
random_indices = np.random.choice(len(table_galaxies), size=20, replace=False)
mini_catalogue = table_galaxies[random_indices]

# Save the mini-catalogue to a new FITS file
mini_catalogue.write("mini_cat_20galaxiesonly.fits", format="fits", overwrite=True)

print("Mini-catalogue saved as 'mini_cat_20galaxiesonly.fits'")
```
